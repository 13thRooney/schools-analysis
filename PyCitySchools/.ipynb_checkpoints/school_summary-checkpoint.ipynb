# SCHOOLS ANALYSIS

* As a whole, schools with higher budgets, did not yield better test results. By contrast, schools with higher spending per student actually (\$645-675) underperformed compared to schools with smaller budgets (<\$585 per student).

* As a whole, smaller and medium sized schools dramatically out-performed large sized schools on passing math performances (89-91% passing vs 67%).

* As a whole, charter schools out-performed the public district schools across all metrics. However, more analysis will be required to glean if the effect is due to school practices or the fact that charter schools tend to serve smaller student populations per school.

```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas Data Frames
schools_df = pd.read_csv(school_data_to_load)
students_df = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset
combined_df = pd.merge(students_df, schools_df, how="left", on=["school_name", "school_name"])
combined_df.head()
```

## District Summary

* Calculate the total number of schools

* Calculate the total number of students

* Calculate the total budget

* Calculate the average math score 

* Calculate the average reading score

* Calculate the overall passing rate (overall average score), i.e. (avg. math score + avg. reading score)/2

* Calculate the percentage of students with a passing math score (70 or greater)

* Calculate the percentage of students with a passing reading score (70 or greater)

```python
# Total number of schools
total_schools = schools_df["school_name"].nunique()
total_schools
```

```python
# Total number of students
total_students = int(students_df["Student ID"].nunique())
total_students
```

```python
# Total Budget
total_budget = schools_df["budget"].sum()
total_budget
```

```python
# Average Math Score
avg_math_score = combined_df["math_score"].mean()
avg_math_score
```

```python
# Average Reading Score
avg_reading_score = combined_df["reading_score"].mean()
avg_reading_score
```

```python
# Calculate the overall passing rate (overall average score), i.e. (avg. math score + avg. reading score)/2
overall_pass_rate = (avg_math_score + avg_reading_score)/2
overall_pass_rate
```

```python
#pct_passing_math = #if math_score >=70
math_score = combined_df["math_score"]
math_score

total_passed_math = sum(math_score >= 70)
total_passed_math

pct_passed_math = (total_passed_math / total_students) * 100 
pct_passed_math
```

```python
#pct_passing_reading = #if reading_score >=70

reading_score = combined_df["reading_score"]
reading_score

total_passed_reading = sum(reading_score >= 70)
total_passed_reading

pct_passed_reading = (total_passed_reading / total_students) * 100 
pct_passed_reading
```

```python
#Create a dataframe to hold the above results
district_summary_df = pd.DataFrame({
    "Total Schools": [total_schools],
    "Total Students": [total_students],
    "Total Budget": [total_budget],
    "Average Math Score": [avg_math_score],
    "Average Reading Score": [avg_reading_score],
    "% Passing Math": [pct_passed_math],
    "% Passing Reading": [pct_passed_reading],
    "% Overall Passing Rate": [overall_pass_rate]
})
district_summary_df
```

```python
#Optional: give the displayed data cleaner formatting
district_summary_df["Total Students"] = district_summary_df["Total Students"].map("{:,}".format)
district_summary_df["Total Budget"] = district_summary_df["Total Budget"].map("${:,}".format)
district_summary_df["Average Math Score"] = district_summary_df["Average Math Score"].map("{:.2f}".format)
district_summary_df["Average Reading Score"] = district_summary_df["Average Reading Score"].map("{:.2f}".format)
district_summary_df["% Passing Math"] = district_summary_df["% Passing Math"].map("{:.2f}%".format)
district_summary_df["% Passing Reading"] = district_summary_df["% Passing Reading"].map("{:.2f}%".format)
district_summary_df["% Overall Passing Rate"] = district_summary_df["% Overall Passing Rate"].map("{:.2f}%".format)
district_summary_df
```

## School Summary

* Create an overview table that summarizes key metrics about each school, including:
  * School Name
  * School Type
  * Total Students
  * Total School Budget
  * Per Student Budget
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * Overall Passing Rate (Average of the above two)
  
* Create a dataframe to hold the above results

```python
# Map type of school to school name

school_type = schools_df.set_index(["school_name"])["type"]
school_type
```

```python
# Total Students per School

students_per_school = combined_df["school_name"].value_counts()
students_per_school
```

```python
# Total Budget per School

budget_per_school = combined_df.groupby(["school_name"]).mean()["budget"]
budget_per_school
```

```python
# Per Student Budget

per_student_budget = budget_per_school / students_per_school
per_student_budget
```

```python
#Average Math Scores

avg_math_score = combined_df.groupby(["school_name"]).mean()["math_score"]
avg_math_score
```

```python
#Average Reading Scores

avg_reading_score = combined_df.groupby(["school_name"]).mean()["reading_score"]
avg_reading_score
```

```python
# Percentage of students passing math by school

students_passing_math = combined_df[(combined_df["math_score"] >= 70)]
students_passing_math_per_school = students_passing_math.groupby(["school_name"]).count()["student_name"]
pct_students_passing_math_by_school = (students_passing_math_per_school / students_per_school) * 100
pct_students_passing_math_by_school
```

```python
# Percentate of students passing reading by school

students_passing_reading = combined_df[(combined_df["reading_score"] >= 70)]
students_passing_reading_per_school = students_passing_reading.groupby(["school_name"]).count()["student_name"]
pct_students_passing_reading_by_school = (students_passing_reading_per_school / students_per_school) * 100
pct_students_passing_reading_by_school
```

```python
# Overall Passing Rate

overall_passing_rate = (pct_students_passing_math_by_school + pct_students_passing_reading_by_school) / 2
overall_passing_rate
```

```python
# Create a school_summary dataframe to hold the above results
school_summary_df = pd.DataFrame({
    "School Type": school_type,
    "Total Students": students_per_school,
    "Total School Budget": budget_per_school,
    "Per Student Budget": per_student_budget,
    "Average Math Score": avg_math_score,
    "Average Reading Score": avg_reading_score,
    "% Passing Math": pct_students_passing_math_by_school,
    "% Passing Reading": pct_students_passing_reading_by_school,
    "% Overall Passing Rate": overall_passing_rate
})
school_summary_df
```

## Top Performing Schools (By Passing Rate)

```python
# Top Performing Schools (By Passing Rate)

top_performing_schools = school_summary_df.sort_values(["% Overall Passing Rate"],ascending = False).head(5)
top_performing_schools
```

## Scores by School Spending

* Create a table that breaks down school performances based on average Spending Ranges (Per Student). Use 4 reasonable bins to group school spending. Include in the table each of the following:
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * Overall Passing Rate (Average of the above two)

```python
# Scores by School Spending

spending_bins = [0, 585, 615, 645, 675]
group_names = ["<$585", "$585-615", "$615-645", "$645-675"]

school_summary_df["Spending Ranges (Per Student)"] = pd.cut(per_student_budget, spending_bins, labels=group_names)

scores_by_school_df = school_summary_df.groupby("Spending Ranges (Per Student)").mean()

organized_scores_by_spending_df = scores_by_school_df[["Average Math Score", "Average Reading Score", "% Passing Math", "% Passing Reading", "% Overall Passing Rate"]]
organized_scores_by_spending_df
```

## Scores by School Size

* Perform the same operations as above, based on school size.

```python
# Scores by School Size

size_bins = [0, 1000, 2000, 5000]
group_names = ["Small (<1000)", "Medium (1000-2000)", "Large (2000-5000)"]

school_summary_df["School Size"] = pd.cut(students_per_school, size_bins, labels=group_names)

scores_by_school_size_df = school_summary_df.groupby("School Size").mean()

organized_scores_by_school_size_df = scores_by_school_size_df[["Average Math Score", "Average Reading Score", "% Passing Math", "% Passing Reading", "% Overall Passing Rate"]]
organized_scores_by_school_size_df
```
